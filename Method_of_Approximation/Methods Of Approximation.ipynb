# Number Of Positive/Negative Roots

```python
def signsCal(coef):
    signs = []
    for i in coef:
        if i >= 0:
            signs.append('+')
        else:
            signs.append('-')
    return signs

def nOfChanges(signs):
    c = 0
    for i in range(len(signs)-1):
        if signs[i+1] != signs[i]:
            c += 1
    return c
```

```python
deg = int(input('Enter the Degree of the Polynomial: '))
coef = [float(input(f"Enter the coeffient of x^{deg - i}: ")) for i in range(deg + 1)]

# Number of Positive Roots
signs = signsCal(coef)
print(f'\nThe Number of Real Positive Roots is almost {nOfChanges(signs)}')

# Number of Negative Roots
neg_coef = []

for i in range(len(coef)-1,-1,-1):
    if i % 2 != 0:
        neg_coef.append(coef[len(coef) - i -1]*-1)
    else:
        neg_coef.append(coef[len(coef) - i -1])

print(f'The Number of Real Negative Roots is almost {nOfChanges(signsCal(neg_coef))}')

print(coef)
```

# Number Of Real/Complex Roots (Strum Seq.)

```python
def coef_2_str(rx):
    rem = []
    for i in range(len(rx)-1, -1,-1):
        r1 = ''
        if rx[i] >= 0:
            r1 = '+ '
        if i == 0:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}'
        elif i == 1:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}*x'
        elif i != 0:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}*x**{i}'
        rem.append(r1)
    return rem

def Remainder(rx):
    rem = []
    for i in range(len(rx)-1, -1,-1):
        r1 = ''
        if rx[i] >= 0:
            r1 = '+ '
        if i == 0:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}'
        elif i == 1:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}*x'
        elif i != 0:
            r1 = r1 + f'{Fraction(float(rx[i])).limit_denominator()}*x**{i}'
        rem.append(r1)
    return rem
```

```python
deg = int(input('Enter the Degree of the Polynomial: '))
coef = [float(input(f"Enter the coeffient of x^{deg - i}: ")) for i in range(deg + 1)]
upto = 4

x = symbols('x')

# Converting the Coeffient to Equation form
signs = signsCal(coef)
expr = ''
for i in range(len(coef)):
    if coef[i] >= 0:
        expr += f'{signs[i]}{coef[i]}*x**{int(deg - i)}'
    else:
        expr += f'{coef[i]}*x**{int(deg - i)}'

# Converting to symboic form
temp_num = sympy.parsing.sympy_parser.parse_expr(expr)

# Finding the Roots between the Interval
print("\nFor Infinity Press enter..... ")
while True:
    x_0 = input('Enter the Interval 1: ')
    x_1 = input('Enter the Interval 2: ')
    k = 1e10
    
    if x_0 != '':
        if round(temp_num.subs(x,x_0), upto) != 0:
            x_0 = float(x_0)
        else:
            print(f'You Have Entered The Exact Root of {temp_num} at x = {x_0}\n')
            continue
    else:
        x_0 = -k
        
    if x_1 != '':
        if round(temp_num.subs(x,x_1), upto) != 0:
            x_1 = float(x_1)
        else:
            print(f'You Have Entered The Exact Root of {temp_num} at x = {x_1}\n')
            continue
    else:
        x_1 = k
  
    if (x_0 < x_1):
            break

    print('!! Interval 1 should be less then Interval 2 !!\n')    
    
# Finding the Denominator
temp_den = Derivative(temp_num,x).doit()

# Collecting the remainders
result_rem = [str(temp_num), str(temp_den)]

# Finding the Coeffient of Denominator
coef_den = []
for i in range(deg -1, -1,-1):
    coef_den.append(round(float(temp_den.coeff(x,n = i)), 5))

count = 2
while True:
    if len(coef_den) <= 1:
        break
        
# Reversing the Order of Nr & Dr
    px = coef[::-1]
    gx = coef_den[::-1]

    # Dividing the Two Polynomials
    qx, rx = numpy.polynomial.polynomial.polydiv(px,gx)
    r = rx
    #     quo = []
    #     for i in range(len(qx)-1,-1,-1):
    #         q1 = f'{Fraction(float(qx[::-1][i])).limit_denominator()} * x**{len(qx)-1-i}'
    #         quo.append(q1)

    # Extracting the Remainder for the next Iteration
    result_rem.append(" ".join(Remainder(rx*-1)))
    
    # ReAssigning the coeffients
    coef = coef_den
    coef_den = (rx * -1)[::-1]
    count += 1
    
print('\n')
print('----'*20)

if x_0 == -k:
    x_0 = '-inf'

if x_1 == k:
    x_1 = 'inf'
print(f'|{"Strum Sequence is".center(33)}',f'|{f"x = {x_0}".center(19)}',f'|{f"x = {x_1}".center(20)}|')
print(f'----'*20)

c_r = []; c_c = []
for i in range(len(result_rem)):
    if x_0 == '-inf':
        t_1 = sympy.parsing.sympy_parser.parse_expr(result_rem[i]).subs(x,-k)
    else:
        t_1 = sympy.parsing.sympy_parser.parse_expr(result_rem[i]).subs(x,x_0)
    
    if x_1 == 'inf':
        t_2 = sympy.parsing.sympy_parser.parse_expr(result_rem[i]).subs(x,k)
    else:
        t_2 = sympy.parsing.sympy_parser.parse_expr(result_rem[i]).subs(x,x_1)
        
    if t_1 >= 0: 
        sig1 = '+'
    else: 
        sig1 = '-'
    c_r.append(sig1)
    
    if t_2 >= 0: 
        sig2 = '+'
    else: 
        sig2 = '-'
    c_c.append(sig2)
    
    z = " = " + str(result_rem[i])
    if len(z) > 34:
        z = f"To Big to fit here"
    print(f"| f_{i}(x) {z.ljust(26)}|{sig1.center(20)}|{sig2.center(20)}|")
print('----'*20)

print(f"  \n∴ Number of Real Roots of the Equation f(x) = {result_rem[0]}\n\t\ton [{x_0}, {x_1}] is {nOfChanges(c_r)} - {nOfChanges(c_c)} = {nOfChanges(c_r) - nOfChanges(c_c)}")

print(f'\n\n\tThis are the remainders {result_rem}')
```

## Iterative Method

```python
def f(x):
    return 2 * sin(x)
```

```python
x_0 = float(input("Enter your Initial Guess: "))
precision = 1e-4

c = 0
n = 30
upto = 3
print('-'*n)

while True:
    x_1 = round(f(x_0), upto)
    c += 1
#     print(x_1,f(x_0))
    if abs(x_1 - x_0) < precision or c >= 1000:
        if c>=1000: print("The Loop ended forcefully.......")
        break
#     print(f'x_{c} = {x_1}')
#     print('-'*n)

    x_0 = x_1
    
print(f'With Precision value of {precision}, the approximate value of root of f(x) = 0 is {round(x_0, upto)}, and the loop has run {c} times.')
```

## Regular False Position/ Regula Falsi

```python
def g(x):
    try:
        return x**3 - 2 * x - 5
    except:
        return 0
    
precision = 1e-4

n = 30
x_0 = 0
a, b = 2, 3
c = 1
upto = 2
# print(g(a),g(b))

print('\n')
while True:
#     print(f'a: {a}, b: {b}, g(a): {g(a)}, g(b): {g(b)}')
    
    x_1 = round((a * g(b) - b * g(a))/(g(b) - g(a)), upto)
#     print(f'({a} * {g(b)} - {b} * {g(a)} / ( {g(b)} - {g(a)} )')
    
    if ((g(x_1) > 0 and x_1 < 0) or (g(x_1) < 0 and x_1 > 0)):
        a = x_1
        print(f"In the next iteration a = {x_1}")
    else:
        b = x_1
        print(f'In the next iteration b = {x_1}')
    
    if abs(x_1 - x_0) < precision:
        break
    
    print(f'x_{c} = {x_1}, g({x_1}) = {round(g(x_1), upto)}')
        
    x_0 = x_1
    c += 1
    
    print('-'*n)

print(f'\nThe Approximate root of x is {round(x_1,4)}, i.e g({x_1}) = {round(g(x_1), 3)} ~ 0 and the loop ran {c+1} times, with precision of {precision}')
```

## Secant

```python
def check(x_0, x_1,f):
    return (f(x_0) > 0 and f(x_1) < 0) or (f(x_0) < 0 and f(x_1) > 0)
```

```python
def f(x):
    return x**3 - 5 * x + 1

x = []
y = []

x_0 = 0 #float(input("Enter the initial Value: "))
x_1 = 1 #float(input("Enter the initial Value: "))

while True:
    if check(x_0, x_1,f):
        break
    else:
        print('In order to use Secant Method the Intermediate value theorem should be satisfied.')
        x_0, x_1 = float(input("Enter the New Value: ")), float(input("Enter the new Value: "))
        check(x_0, x_1,f)
        print('\n')

print(round(f(x_0), 5),round(f(x_1), 5))
x.append(x_0)
y.append(f(x_0))

precision = 1e-4
c = 0

while True:
#     print(f'x_0: {x_0}, x_1: {x_1}, f(x_1): {f(x_1)}, f(x_0): {f(x_0)}')
    x_2 = (x_0 * f(x_1) - x_1 * f(x_0)) / (f(x_1) - f(x_0))

    if abs(x_2 - x_1) < precision:
        break
    
    x.append(x_2)
    y.append(f(x_2))
    print(f'{c} ({round(x_2, 5)}, {round(f(x_2), 5)})')
    print('---'*10)
    x_0, x_1 = x_1, x_2
    c+=1

print(f'\nThe Approximate root of x is {x_2}, i.e f({x_2}) = {round(f(x_2), 6)} ~ 0 and the loop ran {c+1} times, with precision of {precision}')
```

## Newton-Raphson

```python
def theta(x):
    return x**4 - x - 10

def d_theta(x):
    return 4*x**3 - 1
```

```python
x_0 = 1.5
c = 1
# print(round(theta(x_0), 5), round(d_theta(x_0), 5))
print('x_0: ',x_0)

while True:
    
    x_1 = x_0 - theta(x_0)/d_theta(x_0)

    if abs(x_1 - x_0) < 1e-10:
        break

    print(f'x_{c}:', round(x_1, 5),f', f(x_{c}) = {round(theta(x_1), 5)}, f\'(x_{c}) = {round(d_theta(x_1), 6)}')
    x_0 = x_1
    c+=1
        
print(f'The approximate root of f(x) is {x_1}')
theta(x_1)
```

## Chebyshev's

```python
def f(x, upto = upto):
    return round(x**4 - x - 10, upto)

def d_f(x, upto = upto):
    return round(4 * x**3 - 1, upto)

def d_2_f(x, upto = upto):
    return round(12 * x**2, upto)
```

```python
x_0 = 1.5

c = 0
precision = 1e-20
tn = 7 # Adjusting the width
upto = 3


print(f' n\tx_n\t f(x_n)\t    f\'(x_n)\tf\'\'(x_n)')
print('---'*20)
print(str(c).ljust(4),'|',str(x_0).ljust(tn),'|',str(f(x_0)).ljust(tn),'|', str(d_f(x_0)).ljust(tn),'|',str(d_2_f(x_0)).ljust(tn),'|')
print('---'*20)

while True:
    x_1 = round(x_0 - f(x_0)/d_f(x_0) - 0.5 * d_2_f(x_0) * (f(x_0)**2 / d_f(x_0)**3), upto)
    if abs(x_1 - x_0) < precision:
        break
        
    print(str(c+1).ljust(4),'|',str(x_1).ljust(tn),'|',str(f(x_1)).ljust(tn),'|', str(d_f(x_1)).ljust(tn),'|',str(d_2_f(x_1)).ljust(tn),'|')
    print('---'*20)
    
    x_0 = x_1
    c+=1
    
    if count==2: break
print('')
print(f"The Approximate Root of the Function is {x_0}, f({x_0}) = {f(x_0)}")
```

## Multi-Point Iteration

```python
def f(x, upto):
    return round(x**4 - x - 10, upto)

def d_f(x, upto):
    return round(4*x**3 - 1, upto)
```

```python
x_0 = 1.5
print('x_0 = ', x_0)
precision = 1e-4
c = 0
upto = 5
while True:
    x_1_t = round(x_0 - f(x_0, upto)/d_f(x_0, upto), upto)
    x_1 = round(x_1_t - f(x_1_t, upto)/ d_f(x_0, upto), upto)

    if abs(x_1 - x_0) < precision:
        break
    
    print(f'f({x_0}) = {f(x_0, upto)}, df({x_0}) = {d_f(x_0, upto)}')
    print(f'f({x_1_t}) = {f(x_1_t, upto)}')
    print(f'x_{c+1}_temp = {x_1_t}, x_{c+1} = {x_1}')     
    x_0 = x_1
    c += 1
    print('--'*5)
    
print(f"The Approximate roots is {x_0}, upto {upto} decimal places.")
```

## Brige-Vieta

```python
def synDiv(coef,x_0, upto = 4, print_table = False):
    temp = 0
    out = []
    temp_list = []
    
    for i in coef:
        temp_list.append(round(temp * x_0, upto))
        temp = temp * x_0 + i
        out.append(round(temp, upto))
        
    if out[-1] == 0:
        print(f'{x_0} is the exact root of the given equation')
    
    if print_table:
        print('\n')
        print(f'\t |   ', "\t".join([str(i).rjust(8) for i in coef]))
        print("\t |   ")
        print(f" {str(x_0).ljust(8)}|   ","\t".join([str(i).rjust(8) for i in temp_list]))
        print('----'*25)
        print('\t |   ',"\t".join([str(i).rjust(8) for i in out]))
        print('')

    del temp_list, temp
    return out

synDiv(synDiv([1, -2, 4, -1, -7, 5], 1, print_table=True), 1)
```

```python
p = int(input('Enter the Highest Degree of Polynomial: '))

coef = [float(input(f"Enter the coeffient of x^{p - i}: ")) for i in range(p+1)]

x_0 = float(input('Enter the Initial Root: '))

precision = 1e-4
count = 0
upto = 4

while True:
    print(f'Iteration {count+1}')
    out = synDiv(coef,x_0,upto, True)
    out_2 = synDiv(out[:-1],x_0, upto,True)
    
    p_1 = round(x_0 - out[-1]/out_2[-1], upto)
    
    print(f'p_{count+1} = {x_0} - {out[-1]}/{out_2[-1]} = {p_1}')
    if abs(p_1 - x_0) < precision:
        break

    x_0 = p_1
    count += 1
    print('\n')

print(f'The approximate root is {p_1}')
```

## Bairstow

```python
def synDiv_B(coef,p_0, q_0, table = False, upto = 4):
    p_0 = p_0* -1
    q_0 = q_0* -1
    temp, temp_2 = 0,0
    
    out = []
    temp_list = []
    temp_list_2 = []
    
    c = 1
    for i in range(len(coef)):
        temp_list.append(round(temp*p_0, upto))
        temp_list_2.append(round(temp_2*q_0, upto))
        
        temp = round(temp * p_0 + coef[i] + temp_2 * q_0, upto)
        out.append(temp)
        
        if i>0:
            temp_2 = out[-2]
        c+=1

    if table:
        print('\n')
        print(f'\t |   ', "\t".join([str(i).rjust(8) for i in coef])) 
        print("\t |   ")
        print(f" {str(p_0).ljust(8)}|   ","\t".join([str(i).rjust(8) for i in temp_list]))
        print(f"{str(q_0).ljust(8)} |   ", "\t".join([str(i).rjust(8) for i in temp_list_2]))
        print('----'*25)
        print('\t |   ',"\t".join([str(i).rjust(8) for i in out]))
        print('')
    
    del temp_list, temp_list_2
    return out
```

```python
def Qud_roots(a,b,c, upto=5):
    
    if a != 0:
        temp = b**2 - (4 * a * c)
        if temp == 0:
            return -b/(2*a)
        elif temp>0:
            return round((-b + sqrt(temp))/(2*a), upto), round((-b - sqrt(temp))/(2*a), upto)
        else:
            sol1 = (-b + cmath.sqrt(temp))/(2*a)
            sol2 = (-b - cmath.sqrt(temp))/(2*a)
            # From Stackoverflow
            sol1 = round(sol1.real,upto) + round(sol1.imag,upto) * 1j
            sol2 = round(sol2.real,upto) + round(sol2.imag,upto) * 1j
            
            return sol1,sol2
    else:
          raise BaseException("Cannot Divide it by Zero")
```

## Non-Linear System of Equations

```python
def f(x,y,upto=5):
    return round(x**2 - y**2 - 3, upto)

def g(x,y,upto=5):
    return round(x**2 + y**2 - 13, upto)

def df_x(x,y = 0, upto= 5):
    return round(2*x,upto)

def df_y(y, x = 0, upto= 5):
    return round(-2 * y, upto)

def dg_x(x,y = 0, upto= 5):
    return round(2 * x, upto)

def dg_y(y, x = 0,upto=5):
    return round(2 * y, upto)
```

```python
x_0,y_0 = sqrt(6.5), sqrt(6.5)

count = 0
precision = 1e-20
upto = 5

print(f'x_{count} = {x_0}, y_{count+1} =  {y_0}\n{"-----"*8}')

while True:
    
    a_1 = df_x(x_0); b_1 = df_y(y_0); a_2 = dg_x(x_0); b_2 = dg_y(y_0)
    det = (a_1 * b_2 - b_1 * a_2)
    
    if det != 0:
        k = round((-g(x_0, y_0) * a_1 - a_2 * -f(x_0, y_0))/(b_2 * a_1 - b_1 * a_2), upto)
        h = round((-f(x_0, y_0) - k * b_1)/a_1, upto)
        
        x_1 = round(x_0 + h, upto)
        y_1 = round(y_0 + k, upto)
        
   
    if abs(x_1 - x_0) < precision and abs(y_1 - y_0) < precision:
        break
    
    print(f'x_{count+1} = {x_1}, y_{count+1} =  {y_1}')
    x_0 = x_1
    y_0 = y_1
    count += 1
    print('-----'*8)
    
print(f'The Approximate root of f(x) and g(x) is x_0 = {x_0}, y_0 = {y_0}')
print(f'f({x_0}, {y_0}) = {f(x_0,y_0)} & g({x_0},{y_0}) = {g(x_0,y_0)}, upto {upto} decimal points with the precision of {precision}')
```
